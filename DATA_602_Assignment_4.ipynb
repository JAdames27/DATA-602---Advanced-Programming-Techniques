```python
#Julian Adames-Ng
#DATA 602 - Advanced Programming Techniques
#Prof. Schettini

#Q1: Create a class called BankAccount that has four attributes: bankname, firstname, lastname, and balance.
class BankAccount():
  #initialize the attributes. The default balance should be set to 0.
  def __init__(self, bankname, firstname, lastname, balance = 0):
    self.bankname = bankname
    self.firstname = firstname
    self.lastname = lastname
    self.balance = balance

  #● A method called deposit() that allows the user to make deposits into their balance.
  def deposit(self, dep):
    self.balance = dep + self.balance
    return f"You've deposited ${dep} into your account. \nYour balance is now ${self.balance}.\n"

  #● A method called withdrawal() that allows the user to withdraw from their balance.
  #● Withdrawal may not exceed the available balance.
  def withdraw(self, wd):
    #Hint: consider a conditional argument in your withdrawal() method.
    if wd <= self.balance:
      self.balance = self.balance - wd
      return f'You withdrew ${wd}.\nYour remaining balance is ${self.balance}.\n\n'
    #if a withdrawal request is larger than the balance, the following error is returned
    else:
      return 'The requested withdrawal amount exceeds your balance.\n\n'

  #● Use the __str__() method in order to display the bank name, owner name, and current balance.
  def __str__(self):
    return f"The account owner is {self.firstname} {self.lastname}.\nThey currently bank with {self.bankname} and have a balance of ${self.balance}.\n"

#create instances of the BankAccount() class
emp_1 = BankAccount('Chase', 'Jimmy', 'Smith', 1000)
emp_2 = BankAccount('Bank of America', 'Jessie', 'James', 1500)
emp_3 = BankAccount('CitiBank', 'Will', 'Jones', 20000)
emp_4 = BankAccount('TD Bank', 'Amanda', 'Bynes', 480000)

#● Make a series of deposits and withdrawals to test your class.
print(emp_4.deposit(100))
print(emp_4.withdraw(200988))
print(emp_4.deposit(207))
print(emp_4.withdraw(62000))
```

```python
import numpy as np

#Q2: Create a class Box that has attributes length and width that takes values for length and width upon construction (instantiation via the constructor).
class Box:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  #● A method called render() that prints out to the screen a box made with asterisks of length and width dimensions
  def render(self):
    print(self.width * '*')
    if self.length > 1:
      if self.width > 1:
        for i in range(1, self.length - 1):
          print('*' + (self.width-2) * ' ' + '*')
        print(self.width * '*')
      else:
        for i in range(1, self.length):
          print('*')

  #● A method called invert() that switches length and width with each other
  def invert(self):
    x = self.length
    self.length = self.width
    self.width = x
    return self.length, self.width

  #● Methods get_area() and get_perimeter() that return appropriate geometric calculations
  def get_area(self):
    return self.length * self.width

  def get_perimeter(self):
    return 2*self.length + 2*self.width

  #● A method called double() that doubles the size of the box. Hint: Pay attention to return value here.
  def double(self):
    return 2*(self.length * self.width)

  #● Implement __eq__ so that two boxes can be compared using ==. Two boxes are equal if their respective lengths and widths are identical.
  def __eq__(self, other):
    if self.length == other.length and self.width == other.width:
      return True
    return False

  #● A method print_dim() that prints to screen the length and width details of the box
  def print_dim(self):
    print(f"Length: {self.length}\nWidth: {self.width}\n")

  #● A method get_dim() that returns a tuple containing the length and width of the box
  def get_dim(self):
    btuple = (self.length, self.width)
    return btuple

  #● A method combine() that takes another box as an argument and increases the length and width by the dimensions of the box passed in
  def combine(self, other):
    nlength = self.length + other.length
    nwidth = self.width + other.width
    return nlength, nwidth

  #● A method get_hypot() that finds the length of the diagonal that cuts through the middle
  def get_hypot(self):
    return np.sqrt(self.length**2 + self.width**2)

#● Instantiate 3 boxes of dimensions 5, 10, 3, 4 and 5, 10 and assign to variables box1, box2 and box3 respectively
box1 = Box(5, 10)
box2 = Box(3, 4)
box3 = Box(5, 10)

#render()
box1.render()
box2.render()
box3.render()

#invert()
print(f"Inverted Box1: {box1.invert()}")
print(f"Inverted Box2: {box2.invert()}")
print(f"Inverted Box3: {box3.invert()}\n")

#get_area()
print(f"Area1: {box1.get_area()}")
print(f"Area2: {box2.get_area()}")
print(f"Area3: {box3.get_area()}\n")

#get_perimeter()
print(f"Perimeter1: {box1.get_perimeter()}")
print(f"Perimeter2: {box2.get_perimeter()}")
print(f"Perimeter3: {box3.get_perimeter()}\n")

#● Print dimension info for each using print_dim()
box1.print_dim()
box2.print_dim()
box3.print_dim()

#● Evaluate if box1 == box2, and also evaluate if box1== box3, print True or False to the screen accordingly
print(f"{box1 == box3}\n")

#● Combine box3 into box1 (i.e. box1.combine())
print(f"{box1.combine(box3)}\n")

#● Double the size of box2
print(f"{box2.double()}\n")

#● Combine box2 into box1
print(box1.combine(box2))
```
